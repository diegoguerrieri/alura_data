# Importando Pandas e Requests


```python
import pandas as pd
import requests as rq
```

```python
r = rq.get('https://www.alura.com.br/api/cursos')
r.status_code
r.headers['content-type']
```

# Criando *Dataframe*

```python
base = r.json()
base
dados = pd.json_normalize(data = base, max_level= 0)
dados
```

## Detalhes dos Cursos

```python
# r1 = rq.get('https://cursos.alura.com.br/api/curso-dot-net-6-identity-controle-usuario')
# r1.status_code
# r1.headers['content-type']
# base2 = r1.json()
# base2
# dados2 = pd.json_normalize(base2)
# dados2
slug = dados['slug'].unique()
slug
r2 = []
for i in slug:
   r2 += [pd.json_normalize(rq.get('https://cursos.alura.com.br/api/curso-{}'.format(i)).json(), max_level= 0)]
   
r3 = pd.concat(r2)
r3.reset_index(inplace= True)
r3
```

# Removendo colunas

```python
pd.reset_option("display.max_columns")
r3.drop(labels= ['index', 'ementa', 'chamadas', 'publico_alvo', 'video_1a_aula', 'instrutores', 'musicas', 'depoimentos', 'transcriptions', 'highlightedExtraField', 'curso_substituto', 'meta_title', 'video_1a_aula', 'parceiros', 'formacoes'], axis= 1, inplace= True)
r3
```

# Capturando o total de alunos

```python
r3['alunos_ativos_total'] = pd.json_normalize(r3['alunos_ativos'])['total']
r3
```

# Nome da Categoria e Subcategoria

```python
r3['categoria_nome'] = pd.json_normalize(r3['categoria'])['nome']
r3['subcategoria_nome'] = pd.json_normalize(r3['subcategoria'])['nome']
r3
```

# Classificação pela quantidade de alunos

```python
classes = ['Moderadamente Popular', 'Popular', 'Muito Popular']
bins = [r3['quantidade_alunos'].min(),r3['quantidade_alunos'].quantile(0.33), r3['quantidade_alunos'].quantile(0.66), r3['quantidade_alunos'].max()]
bins
r3['popularidade'] = pd.cut(x = r3['quantidade_alunos'], bins = bins, labels= classes, include_lowest= True)
r3
```

# Notas

```python
notas = r3.groupby(by= 'categoria_nome')['nota'].agg(['min', 'max', 'mean', 'std']).rename(columns = {'min': 'Minimo', 'max': 'Maximo', 'mean': 'Media', 'std': 'Desv_p'})
notas.sort_values(by= ['Media','Maximo'], ascending= False, inplace= True)
notas.style.format('{:,.3f}').highlight_max(color= 'lightblue').highlight_min(color = 'lightyellow')
```

# Quantidade de Cursos

```python
cursos = r3.value_counts(subset= ['categoria_nome', 'subcategoria_nome'], normalize= True)
prop_cursos = pd.DataFrame(cursos)
prop_cursos.columns = ['proporcao']
prop_cursos.style.format('{:.2%}').bar(subset = 'proporcao',vmin= 0.0005, color = 'lightgreen')
```
